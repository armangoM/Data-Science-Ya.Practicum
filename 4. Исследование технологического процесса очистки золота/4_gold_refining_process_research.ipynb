### Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Для вычисления используется формула:

$$ Recovery = \cfrac{C(F-T)}{F(C-T)} \times 100 \% $$

Здесь: 
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Очевидно, в качестве параметров *C, F, T* выступают следующие признаки:
- C - **rougher.output.concentrate_au**;
- F - **rougher.input.feed_au**;
- T - **rougher.output.tail_au**.

Вычислим эффективность обогащения на обучающей выборке:

```python
c = train['rougher.output.concentrate_au'] 
f = train['rougher.input.feed_au'] 
t = train['rougher.output.tail_au'] 

recovery = (c * (f - t)) / (f * (c - t)) * 100
```

Пропусков в столбцах нет:

```python
display(c.isna().sum())
display(f.isna().sum())
display(t.isna().sum())
```

```python
recovery.head(5)
```

В столбце **rougher.output.recovery** пропусков также нет, поэтому расчет метрики будет MAE корректен:

```python
train['rougher.output.recovery'].isna().sum()
```

Вычислим MAE между значениями **rougher.output.recovery** и **recovery**:

```python
MAE_recovery = mean_absolute_error(recovery, train['rougher.output.recovery'])
print(MAE_recovery.round(5))
```

```python
plt.figure(figsize=(8, 5))
fig1 = sns.distplot(train['rougher.output.recovery'], kde = False, \
                    bins = np.linspace(start = 0, stop = 100, num = 50), label = 'готовые данные из таблицы') 
fig2 = sns.distplot(recovery, kde = False, \
                    bins = np.linspace(start = 0, stop = 100, num = 50), label = 'данные, рассчитанные по формуле')  
fig2.set_title("Распределение эффективности обогащения", fontsize = 14)
fig2.set_ylabel("Количество", fontsize = 10)
fig2.set_xlabel("% эффективности")
plt.legend()
plt.show()
```

Распределение, рассчитанное по формуле, идентично данному в датафрейме. Среднее абсолютное отклонение равно нулю. Таким образом, эффективность обогащения рассчитана корректно.

```python
def freq_in_str(string, array):
# возвращает количество раз, которое строка string встретилась в массиве array

    count = 0
    for i in array:
        if string in i:
            count += 1
    return count
```

### Составление функции для вычисления итоговой метрики sMAPE

Напишем функцию для вычисления итоговой метрики. Промежуточная формула такова:

$$ sMAPE = \cfrac{1}{N} \sum_{i=1}^{N} \cfrac{|y_i - \hat y_i|}{(|y_i| - |\hat y_i|) / 2} \times 100 \% $$

Здесь: 

$
y_i — значение~целевого~признака~для~объекта~с~порядковым~номером~i~в~выборке,~на~которой~измеряется~качество;\\
\hat y_i — значение~предсказания~для~объекта~с~порядковым~номером~i~в~тестовой~выборке;\\
N — количество~объектов~в~выборке.
$

Финальная формула выглядит следующим образом:


$$ sMAPE_{fin} = 0.25 \times {sMAPE(rougher)} + 0.75 \times sMAPE(final)  \$$

```python
def smape(y_true, y_pred):
    s = (abs(y_true - y_pred) / ((abs(y_true) + abs(y_pred)) / 2)).sum() 
    res = s / len(y_true) * 100
    return res 
```

```python
def smape_final(rougher, final):
    return (0.25 * rougher + 0.75 * final)
```

### Обучение моделей

Выделим целевые и нецелевые признаки из выборок, попутно удалив столбец **date**, который в данном случае не имеет особого смысла для построения модели.

```python
features_train = train_mod.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis = 1)
target_train_r = train_mod['rougher.output.recovery']
target_train_f = train_mod['final.output.recovery']

features_test = test_mod.drop(['date', 'rougher.output.recovery', 'final.output.recovery'], axis = 1)
target_test_r = test_mod['rougher.output.recovery']
target_test_f = test_mod['final.output.recovery']
```

Все наши переменные количественные и требуют масштабирования:

```python
scaler = StandardScaler()
scaler.fit(features_train) 

features_train_st = scaler.transform(features_train)
features_test_st = scaler.transform(features_test) 
```

Обучим несколько разных моделей и оценим их качество кросс-валидацией (количество блоков возьмем по умолчанию равным 5). В качестве метрики используем написанную выше функцию **smape**.

```python
models = [LinearRegression(), DecisionTreeRegressor(random_state = 12345), RandomForestRegressor(random_state = 12345), 
          CatBoostRegressor(random_state = 12345, iterations = 5, silent=True)]
```

Запишем результаты работы моделей в отдельный датафрейм:

```python
models_smape = []

for model in models: 
    scorer = make_scorer(smape, greater_is_better = False) 
    smape_r = cross_val_score(model, features_train_st, target_train_r, scoring = scorer).mean()
    smape_f = cross_val_score(model,  features_train_st,  target_train_f, scoring = scorer).mean()
    models_smape.append({'model': model.__class__.__name__, 'sMAPE_r': smape_r, 'sMAPE_f': smape_f,
                              'sMAPE': smape_final(smape_r, smape_f)})               
pd.DataFrame(models_smape)
```

Наиболее близкое к нулю значение итоговой метрики получено с помощью модели **CatBoostRegressor**. 

Напишем новую функцию, которая посчитает значение итоговой метрики для целевого вектора, состоящего из двух переменных:

```python
def smape_v2(y_true, y_pred):

    s_rougher = (abs(y_true[:,0] - y_pred[:,0]) / ((abs(y_true[:,0]) + abs(y_pred[:,0])) / 2)).sum()
    s_final = (abs(y_true[:,1] - y_pred[:,1]) / ((abs(y_true[:,1]) + abs(y_pred[:,1])) / 2)).sum()
    
    smape_r = s_rougher / len(y_true) * 100
    smape_f = s_final / len(y_true) * 100
    
    smape_fin = 0.25 * smape_r + 0.75 * smape_f
    return smape_fin 
```

```python
features_train_v2 = np.array(features_train_v2)
features_test_v2 = np.array(features_test_v2)
target_train_v2 = np.array(target_train_v2)
target_test_v2 = np.array(target_test_v2)
```

```python
models_v2 = [LinearRegression(), DecisionTreeRegressor(random_state = 12345), RandomForestRegressor(random_state = 12345)]
```

```python
models_smape_v2 = []

for model in models_v2: 
    scorer = make_scorer(smape_v2, greater_is_better = False) 
    smape_v2_fin = cross_val_score(model, features_train_st_v2, target_train_v2, scoring = scorer).mean()
    models_smape_v2.append({'model': model.__class__.__name__, 'sMAPE': smape_v2_fin})   
```

Попробуем подобрать гиперпараметры моделей дерева решений, случайного леса и CatBoost, чтобы получить более оптимальные значения метрики. Также проверим, какой результат покажет линейная регрессия.

Линейная регрессия:

```python
model_lr = LinearRegression()

model_lr.fit(features_train_st, target_train_r)
pred_r = model_lr.predict(features_test_st)

model_lr.fit(features_train_st, target_train_f)
pred_f = model_lr.predict(features_test_st)

smape_r = smape(target_test_r, pred_r)
smape_f = smape(target_test_f, pred_f)

print('sMAPE линейной регрессии на тестовой выборке:', smape_final(smape_r, smape_f).round(7))
```

```python
best_smape = 100
best_depth = 0
for depth in range(1, 26):
    model_dt = DecisionTreeRegressor(max_depth = depth, random_state = 12345) 
    
    model_dt.fit(features_train_st, target_train_r) 
    pred_r = model_dt.predict(features_test_st)

    model_dt.fit(features_train_st, target_train_f)
    pred_f = model_dt.predict(features_test_st)
        
    smape_r = smape(target_test_r, pred_r)
    smape_f = smape(target_test_f, pred_f) 
        
    result = smape_final(smape_r, smape_f)

    if result < best_smape:
        best_smape = result 
        best_depth = depth
            
print('sMAPE наилучшей модели решающего дерева на тестовой выборке:', best_smape.round(7)) 
print("Значение глубины:", best_depth)
```

Случайный лес:

```python
best_smape = 100
best_n_est = 0
best_depth = 0
for est in range(1, 16):
    for depth in range(1, 6):
        model_rf = RandomForestRegressor(random_state = 12345, n_estimators = est, max_depth = depth) 
        
        model_rf.fit(features_train_st, target_train_r) 
        pred_r = model_rf.predict(features_test_st)

        model_rf.fit(features_train_st, target_train_f)
        pred_f = model_rf.predict(features_test_st)
        
        smape_r = smape(target_test_r, pred_r)
        smape_f = smape(target_test_f, pred_f) 
        
        result = smape_final(smape_r, smape_f)
        
        if result < best_smape:
            best_smape = result 
            best_n_est = est
            best_depth = depth

print("sMAPE наилучшей модели случайного на тестовой выборке:", best_smape.round(7))
print("Число деревьев:", best_n_est)
print("Значение глубины:", best_depth)
```

Опытным путем было проверено, что лучший результат при разных значениях количества итераций достигается при глубине дерева, равной 4. Кроме того, метрика начинает ухудшаться после 71-й итерации.
